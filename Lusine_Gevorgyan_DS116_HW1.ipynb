**Part 1**

Load the dataset. Check the first 5 rows.

```python
import pandas as pd
crime_df = pd.read_csv('crime_data.csv')
print(crime_df.head())
```

Identify columns with missing values and their respective counts. Drop columns where more than 50% of the data is missing (store this version as a new dataset).

```python
missing_values = crime_df.isnull().sum()
print("Missing values per column:\n", missing_values)
threshold = len(crime_df) * 0.5
crime_df_cleaned = crime_df.dropna(axis=1, thresh=threshold)
print("\nColumns after dropping:\n", crime_df_cleaned.columns)
```

Convert the DATE OCC column to a datetime format. Extract the year, month, and day into separate columns. Create a new column for the hour using the TIME OCC column.

```python
crime_df_cleaned = crime_df_cleaned.copy()
crime_df_cleaned['DATE OCC'] = pd.to_datetime(crime_df_cleaned['DATE OCC'])

crime_df_cleaned['Year'] = crime_df_cleaned['DATE OCC'].dt.year
crime_df_cleaned['Month'] = crime_df_cleaned['DATE OCC'].dt.month
crime_df_cleaned['Day'] = crime_df_cleaned['DATE OCC'].dt.day

crime_df_cleaned['Hour'] = crime_df_cleaned['TIME OCC'].apply(lambda x: f"{x:04}"[:2])
print(crime_df_cleaned[['DATE OCC', 'Year', 'Month', 'Day', 'Hour']].head())
```

Filter the dataset for crimes that occurred in 2023. Further filter crimes with the description BURGLARY in the Crm Cd Desc colum

```python
crime_2023 = crime_df_cleaned[crime_df_cleaned['Year'] == 2023]
burglary_2023 = crime_2023[crime_2023['Crm Cd Desc'] == 'BURGLARY']
print(burglary_2023.head())
```

Group the data by AREA NAME and calculate the total number of crimes and the average victim age. Sort the results by total crimes in descending order

```python
crime_grouped = crime_df_cleaned.groupby("AREA NAME")
crime_df_cleaned
```

```python
total_crimes = crime_grouped["DR_NO"].count()
total_crimes
```

```python
average_victim_age = crime_grouped["Vict Age"].mean()
average_victim_age
```

```python
crime_summary = pd.DataFrame({
    "Total_Crimes": total_crimes,
    "Avg_Victim_Age": average_victim_age
})
crime_summary
```

```python
crime_summary = crime_summary.sort_values(by="Total_Crimes", ascending=False)
print(crime_summary)
```

**Part 2**


Find the top 3 most frequent Crm Cd Desc values.

```python
crime_counts = crime_df_cleaned["Crm Cd Desc"].value_counts()
top_3_crimes = crime_counts.head(3)
print(top_3_crimes)
```

Group the data by Hour and count the number of crimes.

```python
crime_by_hour = crime_df_cleaned.groupby("Hour")["Crm Cd Desc"].count()
print(crime_by_hour)
```

Group the data by Vict Sex and calculate: Total crimes, Average victim age.

```python
grouped_by_sex = crime_df_cleaned.groupby('Vict Sex').agg(
    Total_Crimes=('DR_NO', 'count'),  
    Avg_Victim_Age=('Vict Age', 'mean')  
).reset_index()
print(grouped_by_sex)
```

1.
Create a new column, Severity Score, based on the following rules:
•
Assign 5 points if a weapon was used (Weapon Used Cd is not null).
•
Assign 3 points for crimes under BURGLARY.
•
Assign 1 point for all other crimes.
•
Group by AREA NAME and find the total severity score for each area.

```python
crime_df_cleaned['Severity Score'] = 0  


if 'Weapon Used Cd' in crime_df_cleaned.columns:
    crime_df_cleaned.loc[crime_df_cleaned['Weapon Used Cd'].notnull(), 'Severity Score'] += 5

crime_df_cleaned.loc[crime_df_cleaned['Crm Cd Desc'] == 'BURGLARY', 'Severity Score'] += 3


crime_df_cleaned.loc[crime_df_cleaned['Severity Score'] == 0, 'Severity Score'] += 1


severity_by_area = crime_df_cleaned.groupby('AREA NAME')['Severity Score'].sum().reset_index()


severity_by_area_sorted = severity_by_area.sort_values(by='Severity Score', ascending=False)


print(severity_by_area_sorted)
```
